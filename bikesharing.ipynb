```python
day = day[['dteday','holiday','weekday','weathersit','temp','hum','windspeed','cnt','season']]
day
```

```python
day['dteday'] = pd.to_datetime(day['dteday'], format='%Y-%m-%d')
day.dtypes
```

```python
day['temp'] = day['temp'] * 41
day['hum'] = day['hum'] * 100
day['windspeed'] = day['windspeed'] * 67
day
```

```python
day['dteday'] = pd.to_datetime(day['dteday'])

day['weekday'] = day['dteday'].dt.day_name()
```

```python
day.head()
```

```python
day['season'] = day['season'].map({1: 'winter', 2: 'spring', 3:'summer',4 : 'fall'})
day
```

```python
day.to_csv('day_clean.csv')
```

```python
features = ['holiday', 'weathersit', 'temp','hum','windspeed']
features
```

```python
train_df = day[day['dteday'].dt.year == 2011]
```

```python
train_X = train_df[features]
train_X
```

```python
train_y = train_df['cnt']
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

model.fit(train_X, train_y)
```
